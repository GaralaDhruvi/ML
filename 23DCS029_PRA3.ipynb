```python
for x in ['Confirmed', 'Cured', 'Deaths']:
    Q1 = df[x].quantile(0.25)
    Q3 = df[x].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    df = df[(df[x] >= lower) & (df[x] <= upper)]
```

*Practical 3 â€” COVID-19 lockdown effect analysis*

Loads 'covid_india.csv' 

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

fn = 'covid_india.csv'
if os.path.exists(fn):
    df = pd.read_csv(fn, parse_dates=['Date'])
else:
    dates = pd.date_range('2020-01-01','2020-12-31')
    states = ['Maharashtra','Karnataka','Delhi','Tamil Nadu']
    rows=[]
    for s in states:
        base = 120 if s=='Maharashtra' else 30
        for d in dates:
            daily = max(0, int(np.random.poisson(base + 10*np.sin(d.dayofyear/30))))
            rows.append({'Date':d,'State':s,'Confirmed':daily})
    df = pd.DataFrame(rows)

states_to_compare = df['State'].unique()[:2].tolist()
dfp = df[df['State'].isin(states_to_compare)].groupby([pd.Grouper(key='Date',freq='W'),'State'])['Confirmed'].sum().reset_index()

plt.figure(figsize=(10,4))
sns.lineplot(data=dfp, x='Date', y='Confirmed', hue='State', marker='o')
plt.title('Weekly confirmed cases â€” comparison'); plt.show()

# Overlay lockdown periods (example dates)
lockdowns = [('2020-03-25','2020-05-31','Lockdown1-2'), ('2020-06-01','2020-07-31','Unlock1')]
for s in states_to_compare:
    plt.figure(figsize=(10,3))
    tmp = df[df['State']==s].set_index('Date').resample('W')['Confirmed'].sum()
    tmp.plot(marker='o'); plt.title(f'Weekly cases: {s}')
    for a,b,label in lockdowns:
        plt.axvspan(pd.to_datetime(a), pd.to_datetime(b), alpha=0.12)
    plt.show()
```
