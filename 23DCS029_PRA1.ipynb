*Practical 1 â€” Wine dataset EDA*

Load UCI Wine dataset and perform head(), info(), describe(), missing-value check, pairplot (subset) and histograms.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine

wine = load_wine(as_frame=True)
df = wine.frame.copy()
df['target'] = df['target']

print("Shape:", df.shape)
display(df.head())
print("\nInfo:")
display(df.info())
print("\nDescribe:")
display(df.describe())

print("\nMissing values per column:")
print(df.isnull().sum())

# Pairplot of selected features (faster)
subset = df[['alcohol','malic_acid','ash','alcalinity_of_ash','target']]
sns.pairplot(subset, hue='target'); plt.show()

# Histograms of features
df.drop(columns='target').hist(bins=12, figsize=(12,8)); plt.tight_layout(); plt.show()
```
